## Univariate Analysis
We select one feature at a time and put them on different data frame


```python
df_setosa = df[df['Species'] == 'Iris-setosa']
df_versicolor = df[df['Species'] == 'Iris-versicolor']
df_virginica = df[df['Species'] == 'Iris-virginica']
```

Plot the 3 different categories using matplot based on the PetalLengthCm for X axis and for the Y axis all records will zero so we can have just one line for the 3 categories.

```python
plt.plot(df_setosa['SepalLengthCm'], np.zeros_like(df_setosa['SepalLengthCm']), 'o')
plt.plot(df_versicolor['SepalLengthCm'], np.zeros_like(df_versicolor['SepalLengthCm']), 'o')
plt.plot(df_virginica['SepalLengthCm'], np.zeros_like(df_virginica['SepalLengthCm']), 'o')
plt.xlabel('Petal Length');
```

## Bivariate Analysis
 for the bivariate analysis, we take 2 features and plot them as a 2d graph. Using seaboarn library that is facetgrid(dataframe, hue='species', size=5).map(plot.scatter,
'petal_length', 'sepal_width').add_legend();

In the plot it's so clear that we can see overlaping between versicolor and virginica

```python
sns.FacetGrid(df, hue='Species', size=5).map(plt.scatter, 'PetalLengthCm', 'SepalWidthCm').add_legend();
```

## Multivariate Analysis
If three features, we can do the plot on on a 3d, but if more than 3 features it's impossible to draw on a 4d, that is why for this purpose we will be using pair plots. We can use seaboard to accomplish this. sns.pairplot(df, hue='species', size=3), then you can drive insights from the plot (patterns and relationships between several variables, and lets us predict the effect change in one variable will have on the other variable)

```python
sns.pairplot(df, hue='Species', size=3);
```
